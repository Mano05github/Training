User-defined- functions

```python
from py4j.protocol import STRING_TYPE
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType

def salary_band(salary):
  if salary >=70000:
    return 'High'
  elif salary >=60000:
    return 'Medium'
  else:
    return 'Low'


salary_band_udf = udf(salary_band,StringType())
df.withColumn('Salary_band',salary_band_udf('salary')).show()
```

```python
# sliderr

display(widgets.IntRangeSlider())
 # only for integer ranges
slider = widgets.IntSlider(
    # value=21,
    max=100,
    min=18,
    description='Age to vote'
)# for any kind of ranges
display(slider)
```

```python
button = widgets.Button(description='Click to see')

def on_the_click(b):
  # return (name.value,'\n',age.value,'\n',slider.value)
  print('Name: ',name.value)
  print('Age: ',age.value)
  print('Slider:',slider.value)

button.on_click(on_the_click)


display(button)
```
