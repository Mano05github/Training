Dataset 3 — CSV (Sales Dataset)

```python
sales_df = spark.read.csv('sales.csv',header=True,inferSchema=True)
sales_df.show()
```

```python
from pyspark.sql.functions import col
sales_df = sales_df.withColumn('revenue', col('quantity') * col('price'))
sales_df.show()
```

```python
from pyspark.sql.functions import sum, avg, desc

total_revenue = sales_df.select(sum('revenue')).collect()[0][0]
print(f"\nTotal Revenue: {total_revenue}")

revenue_per_product = sales_df.groupBy('product').agg(sum('revenue').alias('total_revenue'))
print("\nRevenue per product:")
revenue_per_product.show()

quantity_sold_per_product = sales_df.groupBy('product').agg(sum('quantity').alias('total_quantity_sold'))
print("\nTotal quantity sold per product:")
quantity_sold_per_product.show()
```

```python

best_selling_product = revenue_per_product.orderBy(desc('total_revenue')).limit(1)
print("\nBest selling product (by revenue):")
best_selling_product.show()

employee_revenue = sales_df.groupBy('emp_id').agg(sum('revenue').alias('total_employee_revenue'))
highest_revenue_employee = employee_revenue.orderBy(desc('total_employee_revenue')).limit(1)
print("\nEmployee generating highest revenue:")
highest_revenue_employee.show()

average_sale_value = sales_df.select(avg('revenue')).collect()[0][0]
print(f"\nAverage Sale Value: {average_sale_value}")

products_high_revenue = revenue_per_product.filter(revenue_per_product.total_revenue > 100000)
print("\nProducts generating revenue above 100,000:")
products_high_revenue.show()
```

Dataset 4 — JSON (Orders Dataset)

```python
data_json = spark.read.json('orders.json')
data_json.show()
```

```python
from pyspark.sql.functions import sum, avg, desc, count

print("\nAll Orders:")
data_json.show(data_json.count(), truncate=False)

total_orders = data_json.count()
print(f"\nTotal Orders: {total_orders}")

total_sales_amount = data_json.select(sum('amount')).collect()[0][0]
print(f"\nTotal Sales Amount: {total_sales_amount}")


spending_per_customer = data_json.groupBy('customer').agg(sum('amount').alias('total_spending'))
print("\nTotal Spending Per Customer:")
spending_per_customer.show()

highest_spending_customer = spending_per_customer.orderBy(desc('total_spending')).limit(1)
print("\nHighest Spending Customer:")
highest_spending_customer.show()
```

```python

sales_per_product = data_json.groupBy('product').agg(sum('amount').alias('total_product_sales'))
print("\nTotal Sales Per Product:")
sales_per_product.show()

hyderabad_customers = data_json.filter(data_json.city == 'Hyderabad').select('customer').distinct()
print("\nCustomers from Hyderabad:")
hyderabad_customers.show()


high_value_orders = data_json.filter(data_json.amount > 10000)
print("\nOrders with Amount Greater Than 10,000:")
high_value_orders.show()

orders_per_city = data_json.groupBy('city').count()
print("\nOrders Placed in Each City:")
orders_per_city.show()
```

Final Combined Challenge

```python
res = data.join(sales_df,on = 'emp_id',how = 'inner')
res.show()
```

```python
from pyspark.sql.functions import sum, desc

employee_total_revenue = res.groupBy('emp_id', 'name', 'department', 'salary').agg(sum('revenue').alias('total_revenue'))
print("\nTotal Revenue Generated by Each Employee:")
employee_total_revenue.show()
```

```python
top_5_employees = employee_total_revenue.orderBy(desc('total_revenue')).limit(5)
print("\nTop 5 Employees by Sales:")
top_5_employees.show()
```

```python
department_revenue = res.groupBy('department').agg(sum('revenue').alias('total_department_revenue'))
highest_revenue_department = department_revenue.orderBy(desc('total_department_revenue')).limit(1)
print("\nDepartment Generating Highest Revenue:")
highest_revenue_department.show()
```

```python

final_sales_report_df = employee_total_revenue

final_sales_report_df.coalesce(1).write.option("header", "true").csv("final_sales_report.csv", mode="overwrite")
print("\nFinal sales report saved to 'final_sales_report.csv'")
```
