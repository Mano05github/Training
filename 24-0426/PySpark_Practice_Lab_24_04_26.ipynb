```python
from pyspark.sql.window import Window
from pyspark.sql.functions import rank, dense_rank, row_number

windowSpec = Window.orderBy(desc("price"))
products.withColumn("rank", rank().over(windowSpec)).show()

```

```python
catWindow = Window.partitionBy("category").orderBy(desc("price"))
products.withColumn("rank", rank().over(catWindow)).show()

```

```python
signupWindow = Window.orderBy("signup_date")
customers.withColumn("row_num", row_number().over(signupWindow)).show()

```

```python
custRevWindow = Window.orderBy(desc("total_revenue"))
revenue_df.join(orders, "order_id").groupBy("customer_id").agg(sum("revenue").alias("total_revenue")) \
    .withColumn("rank", rank().over(custRevWindow)).show()

```

```python
revenue_df.join(orders, "order_id").groupBy("customer_id").agg(sum("revenue").alias("total_revenue")) \
    .withColumn("dense_rank", dense_rank().over(custRevWindow)).show()

```

```python
cityWindow = Window.partitionBy("city").orderBy(desc("revenue"))
revenue_df.join(orders, "order_id").join(customers, "customer_id") \
    .withColumn("rank", rank().over(cityWindow)).filter(col("rank") == 1).show()

```

```python
prodCatWindow = Window.partitionBy("category").orderBy(desc("revenue"))
revenue_df.withColumn("rank", rank().over(prodCatWindow)).filter(col("rank") == 1).show()

```

```python
dateWindow = Window.orderBy("order_date").rowsBetween(Window.unboundedPreceding, Window.currentRow)
revenue_df.join(orders, "order_id").withColumn("running_total", sum("revenue").over(dateWindow)).show()

```

```python
prodRunningWindow = Window.partitionBy("product_id").orderBy("order_id").rowsBetween(Window.unboundedPreceding, Window.currentRow)
revenue_df.withColumn("running_total", sum("revenue").over(prodRunningWindow)).show()

```

```python
payWindow = Window.orderBy(desc("amount"))
payments.withColumn("rank", rank().over(payWindow)).show()
```
